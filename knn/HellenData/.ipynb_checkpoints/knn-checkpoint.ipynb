```python
import numpy as np
import operator
```

```python
def fileRead(filename):
    """
    函数说明:打开文件，把文件中的数据进行分类，1代表不喜欢，2代表魅力一般，3代表极具魅力
    Parameters:
            filename:需要解析的文件名
    Return: matReturn 返回特征矩阵
            classVector  返回分类信息
                    
    """
    with open(filename,'r',encoding="utf-8") as f:
        # s.strip(rm)，当rm空时,默认删除空白符(包括'\n','\r','\t',' ')
        fileLines = f.readlines()
        # 如果有BOM，去除文本中的BOM，BOM是微软的标识符（）
        fileLines[0] = fileLines[0].lstrip("\ufeff")
        numberOfLines = len(fileLines)
        matReturn = np.zeros((numberOfLines,3))
        classVector = []
        for i,line in enumerate(fileLines):
            line = line.strip()
            listFromLine = line.split('\t')
            matReturn[i:] = listFromLine[0:3]
            if listFromLine[-1] == 'didntLike':
                classVector.append(1)
            elif listFromLine[-1] == "smallDoses":
                classVector.append(2)
            elif listFromLine[-1] == "largeDoses":
                classVector.append(3)
        
        return matReturn,classVector  
```

```python
mat,classVector = fileRead('datingTestSet.txt')
```

```python
def autoNorm(dataSet):
    """
    函数说明：归一化处理
    Parameters:
            dataSet:特征矩阵
    Return:
            dataSet:经过归一化处理的特征矩阵
            minVals:最小值组合
            rages:变量的最大值和最小值之间的区间大小
    
    """
    # 返回一行，为每列的元素的最小值
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    dataSet = dataSet - minVals
    dataSet = dataSet/ranges    
    return dataSet,minVals,ranges
```

```python
def classify(intX, dataset, labels, k):
    """
    函数说明：KNN算法
    Parameters:
            intX：用于测试的数据
            dataset:训练使用数据
            label:训练分类标签
            k:KNN算法取距离最近的点的个数
    return :
            sortClassCount[0][0] 分类的结果
    """
#     当为0的时候返回的是行数，当为1的时候返回的是列数
    dataSetSize = dataset.shape[0]
#     tile（a,(x,y)）把 数组a叠加为x行，y列
    diffMat = np.tile(intX, (dataSetSize,1))-dataset
    sqrtDiffMat = diffMat**2
    # 行求和，求和完为一列；如果为0则是列求和，得到1行
    sqDistance = sqrtDiffMat.sum(axis =1)
    distance = np.sqrt(sqDistance)
    # 返回的是排序的索引
    sortDistance = distance.argsort()
    classCount = {}
    for i in range(k):
        voteLabel = labels[sortDistance[i]]
        classCount[voteLabel] = classCount.get(voteLabel,0)+1
    #key=operator.itemgetter(1)根据字典的值进行排序
    #key=operator.itemgetter(0)根据字典的键进行排序 
    sortClassCount = sorted(classCount.items(),key=operator.itemgetter(1),reverse=True)
    return sortClassCount[0][0]
```

```python
def main():
    """
    函数说明：调用所有函数，进行分类
    Parameters:

    return :
    """
    # 变量,飞行公里数，游戏消耗时间百分比，每周消耗的冰淇淋公斤数
    inputArray=np.array([40920,8,0.9])
    # 输出结果
    resultList = ["不喜欢","一丢丢喜欢","超级喜欢"]
    mat,classVector = fileRead('datingTestSet.txt')
    mat,minVals, ranges = autoNorm(mat)
    myArray = (inputArray-minVals)/ranges
    classifyResult = classify(myArray,mat,classVector,10)
    print(resultList[classifyResult-1])
    
```

```python
main()
```
